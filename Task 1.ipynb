```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as npy
import random

lapso = 0.001
decaer = 0.1
breackdown = 50
tiempo = npy.arange(0,breackdown,lapso)
x = []
y = []
def Escalamiento(decaer):
    return (decaer * random.uniform(-1,1))
alfa = 1
beta = 0.1
cabo = 0.5
delta = 0.02
x.append(10)
y.append(5) 
for f_i in range(1,len(tiempo)):
    xd = x[f_i-1] * (alfa - beta*y[f_i-1])
    yd = -y[f_i-1]*(cabo - delta*x[f_i-1])
    next_x = x[f_i-1] + xd * lapso
    next_y = y[f_i-1] + yd * lapso
    x.append(next_x + Escalamiento(decaer))
    y.append(next_y + Escalamiento(decaer))

if decaer == 0:
    plt.plot(tiempo, x)
    plt.plot(tiempo, y)
    plt.xlabel('Tiempo')
    plt.ylabel('Presas')
    plt.legend(('Presas', 'Depredadores'))
    plt.title('Sin ruido')
    plt.show()
    plt.plot(x,y)
    plt.xlabel('Depredadores')
    plt.ylabel('Presas')
    plt.title('Sin ruido')
    plt.show()
    
else:
    plt.plot(tiempo, x)
    plt.plot(tiempo, y)
    plt.xlabel('Tiempo')
    plt.ylabel('Presas')
    plt.legend(('Presas', 'Depredadores'))
    plt.title('Task 1')
    plt.show()
    plt.plot(x,y)
    plt.xlabel('Depredadores')
    plt.ylabel('Presas')
    plt.title('Task 1')
    plt.show()
    Ruido = []
    n =[]
    for sdecaerle in range(100):
        Ruido.append(Escalamiento(decaer))
        n.append(sdecaerle)

    plt.plot(n, Ruido)
    plt.xlabel('Ruido')
    plt.ylabel('Ruido')
    plt.title('Task con ruido')
    plt.show()
```
